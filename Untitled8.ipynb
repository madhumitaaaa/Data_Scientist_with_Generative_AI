```python
# SEABORN ASSIGNMENT

import seaborn as sns

# 1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset
# Generate synthetic dataset
np.random.seed(42)  # For reproducibility
x_synthetic = np.random.rand(100) * 100  # Random values between 0 and 100
y_synthetic = 4 * x_synthetic + np.random.normal(0, 50, 100)  # Linear relationship with noise

# Create a scatter plot using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x_synthetic, y=y_synthetic, color="blue", alpha=0.7)
plt.title("Scatter Plot of Synthetic Dataset")
plt.xlabel("x (Synthetic Variable)")
plt.ylabel("y (Synthetic Variable)")
plt.grid(True)
plt.show()
```

```python
# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numeric variable across different categories.
# Generate a dataset with categories and numerical values
np.random.seed(42)  # For reproducibility
categories = ['A', 'B', 'C', 'D', 'E']
data = {
    'Category': np.random.choice(categories, size=200),  # Randomly assign categories
    'Value': np.random.normal(loc=50, scale=10, size=200)  # Numerical values
}
df = pd.DataFrame(data)

# Visualize the distribution of the numerical variable across categories
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df, palette="pastel")
plt.title("Distribution of Numerical Values Across Categories")
plt.xlabel("Category")
plt.ylabel("Value")
plt.show()
```

```python
# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

# Set random seed for reproducibility
np.random.seed(42)

# Generate a synthetic dataset with correlated features
n_samples = 100
mean = [0, 0, 0]
covariance = [[1, 0.8, 0.6],
              [0.8, 1, 0.7],
              [0.6, 0.7, 1]]

# Generate correlated data
data = np.random.multivariate_normal(mean, covariance, n_samples)
df = pd.DataFrame(data, columns=['Feature_1', 'Feature_2', 'Feature_3'])

# Compute correlation matrix
corr_matrix = df.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

# Set random seed for reproducibility
np.random.seed(20)

# Create the dataset for sales data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# Pivot the data to get sales values for each combination of month and day
sales_pivot = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create a heatmap using Plotly
fig = px.imshow(sales_pivot, title="Variation in Sales Across Different Months and Days", color_continuous_scale='YlGnBu')

# Show the plot
fig.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function.

# Create the x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Compute z using the given function
z = np.sin(np.sqrt(x**2 + y**2))

# Prepare the data for the plot
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create a 3D surface plot using Plotly
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Set plot title and axis labels
fig.update_layout(
    title="3D Surface Plot of z = sin(sqrt(x^2 + y^2))",
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

# Show the plot
fig.show()
```

```python
# BOKEH ASSIGNMENT
#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
# Generate data for the sine wave
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add the sine wave as a line
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Show the plot in the notebook
output_notebook()
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
# Generate random data for x, y, sizes, and colors
np.random.seed(42)
n = 100
x = np.random.rand(n) * 10
y = np.random.rand(n) * 10
sizes = np.random.rand(n) * 20 + 5  # Size of the markers
colors = np.random.choice(['red', 'green', 'blue', 'purple'], n)  # Colors for the markers

# Create a ColumnDataSource for Bokeh
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a scatter plot
p = figure(title="Scatter Plot with Different Sizes and Colors", x_axis_label='x', y_axis_label='y', width=800, height=400)

# Add scatter plot with different sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_field='colors', fill_alpha=0.6)

# Show the plot
output_notebook()
show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data
# Generate random data
data_hist = np.random.randn(1000)

# Create histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Create a new plot
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', width=800, height=400)

# Add a quad renderer to the plot
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="skyblue", line_color="white", legend_label="Frequency")

# Show the plot
output_notebook()
show(p)
```

```python
#5. Create a Bokeh heatmap using the provided dataset.
# Generate the heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the grid for Bokeh
x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

# Create a color map for the heatmap
color_mapper = linear_cmap(field_name='z', palette=Viridis256, low=min(z_flat), high=max(z_flat))

# Create the figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', width=800, height=800)

# Add squares to the plot to represent the heatmap
p.square(x='x', y='y', size=25, source=dict(x=x_flat, y=y_flat, z=z_flat),
         color=color_mapper, line_color="white", fill_alpha=0.7)

# Add a color bar to represent the values
color_bar = ColorBar(color_mapper=color_mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
output_notebook()
show(p)
```
